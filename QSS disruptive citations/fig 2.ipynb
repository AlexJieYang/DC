```python
y_pos = [1,2,3,]
d1 = len(fControl[fControl['Field'] == 'Physics'])
d2 = len(fControl[fControl['Field'] == 'Chemistry'])
d3 = len(fControl[fControl['Field'] == 'Medicine'])
d = [d1,d2,d3]
```

```python
f, ax = plt.subplots(figsize=(7, 7))

plt.bar(y_pos, d,align='center', color= ['#212F3D','#212F3D' ,'#212F3D'],alpha=0.9)
ax.yaxis.grid(True)
ax.set_yscale("log")
#plt.ylim(0,300)
plt.xticks(size = 28)
plt.yticks([10,100,1000,10000,100000],size = 28)
plt.xticks([1,2,3],['Physics','Chemistry' , 'Medicine'],fontsize = 25)
plt.ylabel(r'Number',fontsize = 35)
ax.set(xlabel="")
plt.savefig(r"F:\A J YANG PAPER\Nobel analysis\QSS Disruptive citation for Nobel Prize papers\fig\fig 2\b1.pdf",bbox_inches = 'tight')
```

```python
d = fControl['paperyear'].groupby(fControl['paperyear']).count()
```

```python
f, ax = plt.subplots(figsize=(10, 6))
ax.plot(d.index, d.values, linewidth=2.0, c = "#212F3C")
ax.yaxis.grid(True)
plt.xticks(size = 28)
plt.yticks(size = 28)
plt.ylabel(r'Number',fontsize = 35)
plt.xlabel(r'publishing year',fontsize = 35)
plt.savefig(r"F:\A J YANG PAPER\Nobel analysis\QSS Disruptive citation for Nobel Prize papers\fig\fig 2\b2.pdf",bbox_inches = 'tight')
```
