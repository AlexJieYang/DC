```python
import random
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import math
import re
from collections import Counter
import scipy.stats
import seaborn as sns
```

```python
path11 = r'E:\netData\MAGzip\MAG Nobel\2 Nobelpapers\712 NobelpapersCD.zip'
path12 = r'E:\netData\MAGzip\MAG Nobel\2 Nobelpapers\712 Nobelpapers.zip'
path21 = r'E:\netData\MAGzip\MAG Nobel\2 NobelpaperControls\20954 NobelpapercontrolsCD.zip'
path22 = r'E:\netData\MAGzip\MAG Nobel\2 NobelpaperControls\20954 Nobelpapercontrols.zip'
```

```python
fNobel1 = pd.read_csv(path11)
fControl1 = pd.read_csv(path21)
fNobel2 = pd.read_csv(path12)
fControl2 = pd.read_csv(path22)
```

```python
fNobel = pd.concat([fNobel1 , fNobel2] , axis = 1)
fControl = pd.concat([fControl1 , fControl2] , axis = 1)

fNobel = fNobel.loc[:, ~fNobel.columns.duplicated()]
fControl = fControl.loc[:, ~fControl.columns.duplicated()]
```

# fig2

```python
y_pos = [1,2,3,]
d1 = len(fNobel[fNobel['Field'] == 'Physics'])
d2 = len(fNobel[fNobel['Field'] == 'Chemistry'])
d3 = len(fNobel[fNobel['Field'] == 'Medicine'])
d = [d1,d2,d3]

f, ax = plt.subplots(figsize=(10, 7))
plt.bar(y_pos, d,align='center', color= ['#212F3D','#212F3D' ,'#212F3D'],alpha=0.9)
ax.yaxis.grid(True)
plt.ylim(0,300)
plt.yticks(size = 38)
plt.xticks([1,2,3],['Physics','Chemistry' , 'Medicine'],fontsize = 35)
# plt.ylabel(r'Number',fontsize = 35)
ax.set(xlabel="")
ax.set(ylabel="")
plt.savefig(r"G:\A J YANG PAPER\MAG analysis\papers Nobel\QSS Disruptive citation for Nobel Prize papers\fig\fig 2\a1.pdf",bbox_inches = 'tight')
```
